```python
import sqlite3
from IPython.display import display
```

```python
conn=sqlite3.connect('test.db')
df.to_sql("hospital",conn, if_exists="replace",index=False)
```

1. Show all records from the hospital table

```python
query="select * from hospital"
result=pd.read_sql(query,conn)
display(result)
```

2 Count total patients in the table.

```python
query="select count(*) AS total_patients from hospital;"
result=pd.read_sql(query,conn)
display(result)
```

List all unique departments.

```python
query="select distinct department from hospital;"
result=pd.read_sql(query,conn)
display(result)
```

Show all patients whose Gender = 'Male'.


```python
query="select * from hospital where gender = 'Male';"
result=pd.read_sql(query,conn)
display(result)
```

Show patients whose Age is greater than 50.

```python
query="select * from hospital where age > 50;"
result=pd.read_sql(query,conn)
display(result)
```

Show all patients with NULL BillAmount.

```python
query="select * from hospital where billamount is null;"
result=pd.read_sql(query,conn)
display(result)
```

Count how many patients are Admitted (Status = 'Admitted').

```python
query="select count(*) as admitted_patients from hospital where status = 'Admitted';"
result=pd.read_sql(query,conn)
display(result)
```

Show patients from the
Cardiology department.

```python
query="select * from hospital where department = 'Cardiology';"
result=pd.read_sql(query,conn)
display(result)
```

Show patients whose Name starts with 'A'.

```python
query="select * from hospital where name like 'A%';"
result=pd.read_sql(query,conn)
display(result)
```

Show all patients whose AdmissionDate is
NULL.

```python
query="select * from hospital where admissiondate is null;"
result=pd.read_sql(query,conn)
display(result)
```

Show all patients with BillAmount greater than 20000.

```python
query="select * from hospital where billamount > 20000;"
result=pd.read_sql(query,conn)
display(result)
```

Sort all patients by Age descending.

```python
query="select * from hospital order by age desc;"
result=pd.read_sql(query,conn)
display(result)
```

Show the top 5 highest BillAmount records.

```python
query="select * from hospital order by billamount desc limit 5;"
result=pd.read_sql(query,conn)
display(result)
```

Count how many patients were treated by
each doctor (group by Doctor).

```python
query="select doctor, count(*) as total_patients from hospital group by doctor;"
result=pd.read_sql(query,conn)
display(result)
```

Show all rows where PatientID appears more than once.

```python
query="select * from hospital where patientid in ( select patientid from hospital group by patientid having count(*) > 1);"
result=pd.read_sql(query,conn)
display(result)
```
